```python
def summarise_statistics(data, columns):
    summary_stats = {}

    for column in columns:
        # Mean
        mean_value = round(data[column].mean(), 2)
        
        # std dev
        std_value = round(data[column].std(), 2)
        
        summary_stats[column] = {'Mean (M)': mean_value, 'Standard Deviation (SD)': std_value}

    # to DF
    summary_stats_df = pd.DataFrame.from_dict(summary_stats, orient='index')
    
    return summary_stats_df

# Columns in my data (comparison included in 'After' call)
columns_to_analyse = ["Weight", "BMI", "FFM", "VFA"]
```

```python
# Male Control
summary_before_male_control = summarise_statistics(control_male_df, columns_to_analyse)
summary_after_male_control = summarise_statistics(control_male_df, [f"{col} After" for col in columns_to_analyse])


print(summary_before_male_control)
print(summary_after_male_control)
```

```python
# Assumption 1: Test for normal distribution
def shapiro_wilk_test(data, columns):
    normality_results = {}
    for column in columns:
        stat, p = stats.shapiro(data[column])
        normality_results[column] = {
            'Shapiro-Wilk Test Statistic': stat, 
            'p-value': p, 
            'Is Normal': p > 0.05  # Reject null hypo
        }
    return pd.DataFrame.from_dict(normality_results, orient='index')

# Perform normality tests across my data
results_cold_male = shapiro_wilk_test(cold_male_df, columns_to_analyse)
results_control_male = shapiro_wilk_test(control_male_df, columns_to_analyse)
results_cold_female = shapiro_wilk_test(cold_female_df, columns_to_analyse)
results_control_female = shapiro_wilk_test(control_female_df, columns_to_analyse)

print("Normal Distribution Results - Cold Male:")
print(results_cold_male)
print("Normal Distribution Results - Control Male:")
print(results_control_male)
print("Normal Distribution Results - Cold Female:")
print(results_cold_female)
print("Normal Distribution Results- Control Female:")
print(results_control_female)
```

```python
# Assumption 2: Test for Varience
def levene_test(data1, data2, columns, suffix=""):
    variance_results = {}
    for column in columns:
        column_name = column + suffix
        stat, p = stats.levene(data1[column_name], data2[column_name])
        variance_results[column_name] = {
            'Levene Test Statistic': stat, 
            'p-value': p, 
            'Equal Variances': p > 0.05  # Reject null hypo
        }
    return pd.DataFrame.from_dict(variance_results, orient='index')

# Perform varience tests across relative data sets
results_male_variance_before = levene_test(control_male_df, cold_male_df, columns_to_analyse)
results_female_variance_before = levene_test(control_female_df, cold_female_df, columns_to_analyse)
results_male_variance_after = levene_test(control_male_df, cold_male_df, columns_to_analyse, suffix=" After")
results_female_variance_after = levene_test(control_female_df, cold_female_df, columns_to_analyse, suffix=" After")

print("Variance Results - Male Cold (Before):")
print(results_male_variance_before)
print("Variance Results - Female Cold (Before):")
print(results_female_variance_before)
print("Variance Results - Male Control (After):")
print(results_male_variance_after)
print("Variance Results - Female Control (After):")
print(results_female_variance_after)
```

```python
def calculate_p_values_ttest(data, columns):
    p_values = {}

    for column in columns:
        # Before / After
        before_column = column
        after_column = f"{column} After"
        
        if before_column in data.columns and after_column in data.columns:
            # t test
            t_statistic, p_value = ttest_rel(data[before_column].dropna(), data[after_column].dropna())
            
            #P value
            p_values[column] = p_value
        else:
            # None if missing data
            print("Missing Values")
            p_values[column] = None
    
    # to df
    p_values_df = pd.DataFrame(list(p_values.items()), columns=['Variable', 'P-value'])
    
    return p_values_df


def calculate_p_values_wilcoxon(data, columns):
    p_values = {}
    for column in columns:
        #Before / After (Both in the dataframe and named accordingly for ease of use)
        before_column = column
        
        after_column = f"{column} After"
        
        
        if before_column in data.columns and after_column in data.columns:
            stat, p_value = wilcoxon(data[before_column], data[after_column], alternative='two-sided')
            p_values[column] = p_value
        else:
            p_values[column] = None
            print("Missing Values")
    p_values_df = pd.DataFrame(list(p_values.items()), columns=['Variable', 'P-value'])
    return p_values_df
```
